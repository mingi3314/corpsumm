```python
import os
import pandas as pd
import glob
```

```python
for f in glob.glob("results/**/현황 및 전망.html"):
    os.rename(f, f.replace("현황 및 전망", "report"))
```

```python
for f in glob.glob("results/**/사업요약.txt"):
    os.rename(f, f.replace("사업요약", "summary"))
```

```python
def decode_report_url(report_dir):
    with open(report_dir, "r") as f:
        return f.read()

def html_preview(github_html_url):
    # 깃허브에 업로드된 html파일을 렌더링하는 링크 리턴
    src = "http://htmlpreview.github.io/?"
    return src + github_html_url # url encoding

base_url = "https://github.com/mingi3314/corpsumm/blob/main/results"

symbols = os.listdir("results")

df = pd.DataFrame(index = symbols)

for symbol in symbols:
    files = os.listdir(f"results/{symbol}")
    for file in files:
        col = file.split(".")[0]
        ext = file.split(".")[1]

        if '기업분석보고서' in col:
            value = decode_report_url(f"results/{symbol}/{file}")
        elif ext == "html":
            value = f"{base_url}/{symbol}/{file}"
            value = html_preview(value) 
        else:
            value = f"https://raw.githubusercontent.com/mingi3314/corpsumm/main/results/{symbol}/{file}"
        
        df.loc[symbol, col] = value
```
